# IMPORT LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# MEMBACA FILE

```python
original_image = cv2.imread('pcduas.jpg')
```

# KONVERSI CITRA ASLI KE RGB

```python
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
```

# KONVERSI KE GRAYSCALE

```python
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

# MENERAPKAN FILTER MEDIAN

```python
median_filtered_image = np.zeros_like(original_image_rgb)
for i in range(3):
    median_filtered_image[:, :, i] = cv2.medianBlur(original_image_rgb[:, :, i], 3)
```

# MENAMPILKAN CITRA ASLI DAN CITRA FILTER MEDIAN

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
ax = axes.ravel()

ax[0].imshow(original_image_rgb)
ax[0].set_title('Image')
ax[0].axis('on')

ax[1].imshow(median_filtered_image)
ax[1].set_title('Median Filtered Image')
ax[1].axis('on')

plt.show()
```

# KONVERSI KE GRAYSCALE

```python
grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

# MENERAPKAN FILTER MEAN

```python
mean_filtered_image = cv2.blur(grayscale_image, (5, 5))
```

# MENAMPILKAN CITRA GRAYSCALE DAN CITRA FILTER MEAN

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(gray_image, cmap='gray')
ax[0].set_title('Image (Grayscale)')
ax[0].axis('on')

ax[1].imshow(mean_filtered_image, cmap='gray')
ax[1].set_title('Mean Filtered Image')
ax[1].axis('on')

plt.show()
```

# MENAMPILKAN SELURUH CITRA

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0, 0].imshow(original_image_rgb)
axs[0, 0].set_title("image")
axs[0, 0].axis('on')

axs[0, 1].imshow(median_filtered_image)
axs[0, 1].set_title("Median Filtered Image")
axs[0, 1].axis('on')

axs[1, 0].imshow(gray_image, cmap='gray')
axs[1, 0].set_title("Image (Grayscale)")
axs[1, 0].axis('on')

axs[1, 1].imshow(mean_filtered_image, cmap='gray')
axs[1, 1].set_title("Mean Filtered Image")
axs[1, 1].axis('on')

plt.show()
```
